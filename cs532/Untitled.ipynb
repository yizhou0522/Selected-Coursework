```python
# a)
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
in_data = loadmat('face_emotion_data.mat')
y = in_data['y']
X = in_data['X']
```

```python
 w= np.linalg.inv(X.T@X)@X.T@y
print(w)
```

```python
# b) If the new image has a input vector of x, the predicted label for
# this image is sign(x@w). If the result is +1 then the image is a happy face, otherwise
# an angry faces.
```

```python
# c) the first feature vector is more important: since it is 0.94, which is the largest value among all vectors
```

```python
# d)
# easy to find that vectors with greatest values should be used to build classifier
X3 = X[:,[0,2,3]]
w3 = np.linalg.inv(X3.T@X3)@X3.T@y
print(w3)
```

```python
# e)
error = len([1 for i in np.hstack((np.sign(X@w), y)) if i[0]!=i[1]])
print("error rate 9 features: ",error/len(np.sign(X@w))*100,"%")
error_vec_3 = len([1 for i in np.hstack((np.sign(X3@w3), y)) if i[0]!=i[1]])
print("error rate 3 features: ",error_vec_3/len(np.sign(X3@w3))*100,"%")
```

```python
# f)

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
# %%javascript
# IPython.OutputArea.auto_scroll_threshold = 9999

for i in range(8):
 hold = [j for j in range(i*16,i*16+16)]
 error = len([1 for i in np.hstack((np.sign(X[hold,:]@w), y[hold])) if i[0]!=i[1]])
 learn = [j for j in range(len(X)) if j not in hold]
 w = np.linalg.inv(X[learn,:].T@X[learn,:])@X[learn,:].T@y[learn] 
 error=error/len(y[hold])*100  
    
 print(i," Holdout set: ",hold)
 print(i," Training set: ",learn)
 print(i," iteration: error rate ",error,"%")
 print(i," iteration: weight: \n",w)
 print("---------------------------------------------------------------------------------------------")
```
